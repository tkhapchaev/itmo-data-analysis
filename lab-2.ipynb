```python
import implicit
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.sparse import csr_matrix
from sklearn.metrics.pairwise import cosine_similarity
from implicit.nearest_neighbours import ItemItemRecommender
```

```python
items_meta = pd.read_parquet('data\\vk-recsys-challenge\\items_meta.parquet')
train_interactions = pd.read_parquet('data\\vk-recsys-challenge\\train_interactions.parquet')
train_interactions = train_interactions[(train_interactions['like'] + train_interactions['dislike']) >= 1]
users_meta = pd.read_parquet('data\\vk-recsys-challenge\\users_meta.parquet')
test_pairs = pd.read_csv('data\\vk-recsys-challenge\\test_pairs.csv')

print("Дубликаты в users_meta:", users_meta.duplicated().sum())
print("Дубликаты в items_meta:", items_meta.drop(columns=['embeddings']).duplicated().sum())
print("Дубликаты в train_interactions:", train_interactions.duplicated().sum())
print("Дубликаты в test_pairs:", test_pairs.duplicated().sum())

print(f"\nusers_meta ({users_meta.shape[0]}):")
print(users_meta)

print(f"\nitems_meta ({items_meta.shape[0]}):")
print(items_meta)

print(f"\ntrain_interactions ({train_interactions.shape[0]}):")
print(train_interactions)

print(f"\ntest_pairs ({test_pairs.shape[0]}):")
print(test_pairs)
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=users_meta, x='gender')
plt.title("Распределение по полу")
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(users_meta['age'], bins=20, kde=True)
plt.title("Распределение возраста пользователей")
plt.xlabel("Возраст")
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(items_meta['duration'], bins=50, kde=True)
plt.title("Распределение длительности клипов")
plt.xlabel("Длительность (сек)")
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(train_interactions['timespent'], bins=50, kde=True)
plt.title("Распределение времени просмотра (timespent)")
plt.xlabel("Время просмотра")
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.countplot(data=train_interactions, x='like', ax=axes[0,0])
axes[0,0].set_title("Количество лайков")
sns.countplot(data=train_interactions, x='dislike', ax=axes[0,1])
axes[0,1].set_title("Количество дизлайков")
sns.countplot(data=train_interactions, x='share', ax=axes[1,0])
axes[1,0].set_title("Количество репостов")
sns.countplot(data=train_interactions, x='bookmarks', ax=axes[1,1])
axes[1,1].set_title("Количество добавлений в закладки")
plt.tight_layout()
plt.show()
```

- **Распределение по полу**<br/>
Гистограмма показывает, что пользователей с полом «2» значительно больше, чем пользователей с полом «1».
Это говорит о том, что в выборке (или среди аудитории сервиса) преобладает вторая группа.
Вывод: сервис в большей степени представлен пользователями пола «2», что может влиять на контент, который набирает популярность, и на алгоритмы рекомендаций.

- **Распределение возраста пользователей**<br/>
Гистограмма возраста пользователей имеет пик в районе 20–25 лет, затем плавно снижается к 40–50 годам и уходит в хвост к 60+.
Распределение скошено вправо: есть небольшой хвост в сторону старших возрастов.
Наибольшее число пользователей в диапазоне 18–30 лет.
Вывод: аудитория сервиса — преимущественно молодые люди (до 30 лет), но есть и значимое количество пользователей старше 30.

- **Распределение длительности клипов**<br/>
Распределение резко скошено влево: основной пик приходится на 5–10 секунд, чуть меньше в диапазоне до 20–25 секунд.
Небольшие пики присутствуют и в районе 30–40 секунд, но в целом длительность большинства клипов не превышает 20 секунд.
Есть небольшой хвост, доходящий примерно до 180 секунд, но таких клипов очень мало.
Вывод: контент в основном очень короткий (10–20 секунд), что характерно для формата «коротких вертикальных видео». Средняя длительность — явно в районе 10 секунд.

- **Распределение времени просмотра**<br/>
Очень сильный пик на низких значениях (0–10 секунд), затем постепенный спад.
Значительная часть просмотров укладывается в первые 10–20 секунд, потом активность резко падает, но хвост может доходить до 100+ секунд.
Логично, что при длительности ролика 10–20 секунд, часто просмотр будет таким же или меньше.
Вывод: пользователи довольно быстро переключаются на другие видео или закрывают приложение; лишь малая часть досматривает клипы до конца, особенно если клип длиннее 20 секунд.

- **Количество лайков, дизлайков, репостов, добавлений в закладки**<br/>
    - Лайки / дизлайки<br/>
    Лайков очень много, тогда как дизлайков практически не видно на графике.
    Это говорит о сильном дисбалансе между позитивными и негативными реакциями.
    Вывод: пользователи чаще выражают положительную реакцию, а дизлайк ставят реже — возможно, из-за особенностей интерфейса или предпочтений.

    - Репосты<br/>
    Преобладают «0» (то есть не делятся), «1» встречается значительно реже.
    При этом лайков гораздо больше, чем репостов.
    Вывод: репост — более редкое действие, пользователи не так часто делятся клипами.

    - Добавления в закладки<br/>
    Тоже достаточно дисбалансное распределение, но bookmarks (1) встречается чаще, чем репост.
    Однако всё равно в большинстве случаев bookmarks = 0.
    Вывод: добавление в закладки (чтобы вернуться к контенту позже) тоже встречается реже, чем лайк, но чаще, чем репост.



```python
item_popularity = train_interactions.groupby('item_id').agg(
    likes=('like', 'sum'),
    dislikes=('dislike', 'sum')
).reset_index()

N = 100
item_popularity['pop_score'] = item_popularity['likes'].astype('int64') - item_popularity['dislikes'].astype('int64')
top_n_clips = item_popularity.sort_values(by='pop_score', ascending=False).head(N)

print(top_n_clips)
```

```python
most_pop_top_n_recs = test_pairs.copy()

top_n_ids = set(top_n_clips['item_id'])
most_pop_top_n_recs['prediction_mostpop'] = most_pop_top_n_recs['item_id'].isin(top_n_ids).astype(int)

most_pop_top_n_recs.head(most_pop_top_n_recs.shape[0])
```

```python
rec_threshold = 1200
most_pop_with_threshold_recs = test_pairs.copy()

most_pop_with_threshold_recs = most_pop_with_threshold_recs.merge(item_popularity[['item_id', 'pop_score']], on='item_id', how='left')
most_pop_with_threshold_recs['pop_score'] = most_pop_with_threshold_recs['pop_score'].fillna(0)
most_pop_with_threshold_recs['prediction_mostpop'] = (most_pop_with_threshold_recs['pop_score'] > rec_threshold).astype(int)

most_pop_with_threshold_recs.head(most_pop_with_threshold_recs.shape[0])
```

```python
train_interactions_als = train_interactions.copy()
train_interactions_als['rating'] = (
        train_interactions_als['like'].astype('int64') -
        train_interactions_als['dislike'].astype('int64')
)

n_users = train_interactions_als['user_id'].max() + 1
n_items = train_interactions_als['item_id'].max() + 1

rating_matrix = csr_matrix(
    (train_interactions_als['rating'],
     (train_interactions_als['user_id'], train_interactions_als['item_id'])),
    shape=(n_users, n_items)
)

als_model = implicit.als.AlternatingLeastSquares(factors=64, iterations=15, regularization=0.1)
als_model.fit(rating_matrix)

user_factors = als_model.user_factors
item_factors = als_model.item_factors

als_recs = test_pairs.copy()
als_recs['score_als'] = 0
valid_mask = (als_recs['user_id'] < n_users) & (als_recs['item_id'] < n_items)

als_recs.loc[valid_mask, 'score_als'] = np.sum(
    user_factors[als_recs.loc[valid_mask, 'user_id'].values] *
    item_factors[als_recs.loc[valid_mask, 'item_id'].values],
    axis=1
)

als_recs['prediction_als'] = (als_recs['score_als'] > 0).astype(int)
als_recs.head(als_recs.shape[0])
```

```python
train_interactions_itemknn = train_interactions.copy()
train_interactions_itemknn['rating'] = (
        train_interactions_itemknn['like'].astype('int64') -
        train_interactions_itemknn['dislike'].astype('int64')
)

n_users = train_interactions_itemknn['user_id'].max() + 1
n_items = train_interactions_itemknn['item_id'].max() + 1

rating_matrix = csr_matrix(
    (train_interactions_itemknn['rating'].astype(np.float64),
     (train_interactions_itemknn['user_id'], train_interactions_itemknn['item_id'])),
    shape=(n_users, n_items)
)

itemknn_model = ItemItemRecommender(K=1000)
itemknn_model.fit(rating_matrix)

unique_users = test_pairs['user_id'].unique()
N_recs = 100
user_recs = {}
for user in unique_users:
    recs = itemknn_model.recommend(user, rating_matrix, N=N_recs, filter_already_liked_items=False)
    user_recs[user] = { rec[0] for rec in recs }

itemknn_recs = test_pairs.copy()
itemknn_recs['prediction_itemknn'] = itemknn_recs.apply(
    lambda row: 1 if row['item_id'] in user_recs.get(row['user_id'], set()) else 0,
    axis=1
)

print("Пример предсказаний ItemKNN:")
itemknn_recs.head(itemknn_recs.shape[0])
```

```python
pop_df = train_interactions_itemknn.groupby('item_id').agg(count=('rating', 'count')).reset_index()
pop_dict = dict(zip(pop_df['item_id'], pop_df['count']))
total_pop = pop_df['count'].sum()

items_meta_copy = items_meta.copy()
items_meta_copy['embeddings'] = items_meta_copy['embeddings'].apply(lambda emb: np.array(emb))
item_emb_dict = {row['item_id']: row['embeddings'] for idx, row in items_meta_copy.iterrows()}

top_pop_items = set(pop_df.sort_values('count', ascending=False).head(10)['item_id'])

def novelty_for_list(item_list, pop_dict, total_pop):
    scores = []
    for item in item_list:
        count = pop_dict.get(item, 1)
        prob = count / total_pop
        scores.append(-math.log2(prob))
    return np.mean(scores) if scores else np.nan

def diversity_for_list(item_list, item_emb_dict):
    emb_list = [item_emb_dict[item] for item in item_list if item in item_emb_dict]
    if len(emb_list) < 2:
        return 0.0
    embs = np.stack(emb_list)
    sim_matrix = cosine_similarity(embs)
    n = sim_matrix.shape[0]
    triu_indices = np.triu_indices(n, k=1)
    avg_sim = np.mean(sim_matrix[triu_indices])
    return 1 - avg_sim

def unexpectedness_for_list(item_list, top_pop_items):
    if not item_list:
        return np.nan
    unexpected_count = sum(1 for item in item_list if item not in top_pop_items)
    return unexpected_count / len(item_list)

def compute_rec_metrics(rec_df, prediction_col, pop_dict, total_pop, item_emb_dict, top_pop_items):
    recs_by_user = rec_df[rec_df[prediction_col] == 1].groupby('user_id')['item_id'].apply(list)
    novelty_scores = []
    diversity_scores = []
    unexpectedness_scores = []

    for user, rec_list in recs_by_user.items():
        nov = novelty_for_list(rec_list, pop_dict, total_pop)
        novelty_scores.append(nov)
        div = diversity_for_list(rec_list, item_emb_dict)
        diversity_scores.append(div)
        unexp = unexpectedness_for_list(rec_list, top_pop_items)
        unexpectedness_scores.append(unexp)

    avg_novelty = np.nanmean(novelty_scores) if novelty_scores else np.nan
    avg_diversity = np.nanmean(diversity_scores) if diversity_scores else np.nan
    avg_unexpectedness = np.nanmean(unexpectedness_scores) if unexpectedness_scores else np.nan

    return avg_novelty, avg_diversity, avg_unexpectedness

novelty_mostpop, diversity_mostpop, unexpectedness_mostpop = compute_rec_metrics(
    most_pop_top_n_recs, 'prediction_mostpop', pop_dict, total_pop, item_emb_dict, top_pop_items
)

novelty_als, diversity_als, unexpectedness_als = compute_rec_metrics(
    als_recs, 'prediction_als', pop_dict, total_pop, item_emb_dict, top_pop_items
)

novelty_itemknn, diversity_itemknn, unexpectedness_itemknn = compute_rec_metrics(
    itemknn_recs, 'prediction_itemknn', pop_dict, total_pop, item_emb_dict, top_pop_items
)

print("MostPop:")
print("  Новизна:         {:.4f}".format(novelty_mostpop))
print("  Разнообразие:    {:.4f}".format(diversity_mostpop))
print("  Неожиданность:   {:.4f}".format(unexpectedness_mostpop))
print()
print("ALS:")
print("  Новизна:         {:.4f}".format(novelty_als))
print("  Разнообразие:    {:.4f}".format(diversity_als))
print("  Неожиданность:   {:.4f}".format(unexpectedness_als))
print()
print("ItemKNN:")
print("  Новизна:         {:.4f}".format(novelty_itemknn))
print("  Разнообразие:    {:.4f}".format(diversity_itemknn))
print("  Неожиданность:   {:.4f}".format(unexpectedness_itemknn))
```

- **MostPop:**</br>
Новизна (11.48): рекомендации содержат клипы, которые встречались в обучающих данных не очень часто – они обладают определённой новизной, но не экстремальной.</br>
Разнообразие (0.2668): рекомендации по большинству пользователей довольно похожи между собой, что естественно для метода MostPop, который всем выдает один и тот же список самых популярных клипов.</br>
Неожиданность (0.8759): большинство рекомендованных клипов не попадают в глобальный топ-10 – то есть рекомендации «неочевидны».

- **ALS:**</br>
Новизна (16.25): рекомендации от ALS содержат гораздо более «новые» клипы (те, что встречались в обучающих данных ещё реже), что говорит о том, что модель находит менее популярные, но потенциально интересные клипы для пользователей.</br>
Разнообразие (0.8857): рекомендации сильно различаются между собой – это положительный сигнал, так как пользователю предлагается широкий спектр вариантов.</br>
Неожиданность (0.9984): практически все рекомендации – «неожиданные» (не входят в топ-10 по популярности). Это говорит о сильной персонализации, когда модель предлагает клипы, которые в глобальном списке не являются очевидными.

- **ItemKNN:**</br>
Новизна (12.71): рекомендации по новизне чуть выше, чем у MostPop, но ниже, чем у ALS.</br>
Разнообразие (0.1179): это означает, что рекомендуемые клипы для каждого пользователя очень похожи друг на друга.</br>
Неожиданность (0.9939): все рекомендации считаются неожиданными, то есть ни один из рекомендованных клипов не входит в глобальный топ-10.
